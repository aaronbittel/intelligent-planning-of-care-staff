```python
# Standardize dates of D.O.A and D.O.D
DOA = df.columns[2]
DOD = df.columns[3]

def parse_month(month_str):
    match month_str:
        case 'Jan':
            return 1
        case 'Feb':
            return 2
        case 'Mar':
            return 3
        case 'Apr':
            return 4
        case 'May':
            return 5
        case 'Jun':
            return 6
        case 'Jul':
            return 7
        case 'Aug':
            return 8
        case 'Sep':
            return 9
        case 'Oct':
            return 10
        case 'Nov':
            return 11
        case 'Dec':
            return 12
        case _:
            raise ValueError

def parse_date(date_str, month_str):
    try:
        date_time = pd.to_datetime(date_str, format='%m/%d/%Y')
        if date_time.month != parse_month(month_str):
            raise ValueError
        return date_time
    except ValueError:
        try:
            date_time = pd.to_datetime(date_str, format='%d/%m/%Y')
            if date_time.month != parse_month(month_str):
                raise ValueError
            return date_time
        except ValueError:
            return pd.NaT

df['DOA_date_time'] = df.apply(lambda row: parse_date(row[DOA], row['month']), axis=1)

def calc_dod(doa_date_time, duration_of_stay):
    return doa_date_time + pd.DateOffset(days=duration_of_stay-1)
    
df['DOD_date_time'] = df.apply(lambda row: calc_dod(row['DOA_date_time'], row['DURATION OF STAY']), axis=1)

df
```

```python
# Find duplicated entries
is_duplicated = df.drop(columns=['SNO']).duplicated(keep='first')
duplicates = is_duplicated[is_duplicated == True]
duplicates
```

```python
# Remove duplicated entries
df.drop(columns=['SNO'], inplace=True)
df.drop_duplicates(inplace=True)
df
```

```python
def get_date_range(doa, dod):
    return pd.date_range(start=doa, end=dod)
        
df['range'] = df.apply(lambda row: get_date_range(row['DOA_date_time'], row['DOD_date_time']), axis=1)

df_occupancy = df['range'].explode().value_counts()
df_doa = df['DOA_date_time'].value_counts()
df_dod = df['DOD_date_time'].value_counts()
```

```python
# Merge counts into a single data frame
dfm = pd.merge(df_occupancy,df_doa, how='outer', left_index=True, right_index=True)
dfm = pd.merge(dfm,df_dod, how='outer', left_index=True, right_index=True)
dfm.reset_index(drop=False, inplace=True)
dfm.rename(columns={'index': 'dates', 'count_x': 'occupancy', 'count_y': 'admissions', 'count': 'discharges'}, inplace=True)

pd.options.display.max_rows = None
dfm
```

```python
# Plot occupancy by time of a subset of rows [100:300[
dfm[['dates', 'occupancy']][100:300].plot(x='dates', y='occupancy', kind='bar', figsize=(10, 6), xticks=[])
```
